## Part 1: Load in and understand dataset

```python
#import csv library, open insurance.csv, obtain col headings, and read in using csv.DictReader
import csv

ages = []
insurance_full = []

with open('insurance.csv', newline='') as insurance_csv:
    insurance_dataset = csv.DictReader(insurance_csv)
#     for row in insurance_dataset:
#         ages.append(row['age'])
#     print(min(ages), max(ages)) :: min age = 18; max age = 64
#     for row in insurance_dataset:
#         print(row)
    insurance_full = list(insurance_dataset)
print(insurance_full[0:4])
        


```

## Part 2: In what age group is smoking most prevalent? How does that impact insurance cost per age group?

```python
#####################################################################################################################
#####################################################################################################################

# write a function to separate insurance dataset by age group:

def age_sep(dataset):
    
    # create age groups 0-25, 26-35, 36-45, 46-55, 56-65, 66- :
    age_scale = {0: 25, 1: 35, 2: 45, 3: 55, 4: 65}

    # separate insurance dataset based on age according to age_scale:
    age_grouping = {0:[], 1:[], 2:[], 3:[], 4:[], 5:[]}

    for i in range(0, len(dataset)):
        age = int(dataset[i]['age'])
        if age <= age_scale[0]:
            age_grouping[0].append(dataset[i])
        elif age > age_scale[0] and age <= age_scale[1]:
            age_grouping[1].append(dataset[i])
        elif age > age_scale[1] and age <= age_scale[2]:
            age_grouping[2].append(dataset[i])
        elif age > age_scale[2] and age <= age_scale[3]:
            age_grouping[3].append(dataset[i])
        elif age > age_scale[3] and age <= age_scale[4]:
            age_grouping[4].append(dataset[i])
        elif age > age_scale[4]:
            age_grouping[5].append(dataset[i])
    return age_grouping

age_groups = age_sep(insurance_full)
#print(age_groups)
#####################################################################################################################
#####################################################################################################################

# write a function to count the number and proportion of smoker/non-smoker within each age group:

def smoking(dataset):
    # Group 0: Under 25 years old
    smoker0 = 0
    nonsmoker0 = 0

    smoker0ins = 0
    nonsmoker0ins = 0


    for i in range(0,len(dataset[0])):
        if dataset[0][i]['smoker'] == 'no':
            nonsmoker0 += 1
            nonsmoker0ins += float(dataset[0][i]['charges'])
        else:
            smoker0 += 1
            smoker0ins += float(dataset[0][i]['charges'])

    smoker0pct = round(smoker0/len(dataset[0])*100, 2)
    nonsmoker0pct = round(nonsmoker0/len(dataset[0])*100, 2)

    smoker0ins_avg = round(smoker0ins/smoker0, 2)
    nonsmoker0ins_avg = round(nonsmoker0ins/nonsmoker0, 2)
    diff0 = round(smoker0ins_avg - nonsmoker0ins_avg, 2)

    print("There are {smoker}({smokerpct}%) smokers and {nonsmoker}({nonsmokerpct}%) non-smokers in people under 25 years old.".format(smoker=smoker0, smokerpct=smoker0pct, nonsmoker=nonsmoker0, nonsmokerpct=nonsmoker0pct))
    print("In people under 25 years old, the average insurance cost for smokers is ${smokerins} and for non-smokers is ${nonsmokerins} (${diff} difference).".format(smokerins=smoker0ins_avg, nonsmokerins=nonsmoker0ins_avg, diff = diff0))
    print("")

    # Group 1: 26-35 years old
    smoker1 = 0
    nonsmoker1 = 0

    smoker1ins = 0
    nonsmoker1ins = 0

    for i in range(0,len(dataset[1])):
        if dataset[1][i]['smoker'] == 'no':
            nonsmoker1 += 1
            nonsmoker1ins += float(dataset[1][i]['charges'])
        else:
            smoker1 += 1
            smoker1ins += float(dataset[1][i]['charges'])


    smoker1pct = round(smoker1/len(dataset[1])*100, 2)
    nonsmoker1pct = round(nonsmoker1/len(dataset[1])*100, 2)

    smoker1ins_avg = round(smoker1ins/smoker1, 2)
    nonsmoker1ins_avg = round(nonsmoker1ins/nonsmoker1, 2)
    diff1 = round(smoker1ins_avg - nonsmoker1ins_avg, 2)

    print("There are {smoker}({smokerpct}%) smokers and {nonsmoker}({nonsmokerpct}%) non-smokers in people between the ages of 26 and 35".format(smoker=smoker1, smokerpct=smoker1pct, nonsmoker=nonsmoker1, nonsmokerpct=nonsmoker1pct))
    print("In people between 26 and 35 years old, the average insurance cost for smokers is ${smokerins} and for non-smokers is ${nonsmokerins} (${diff} difference).".format(smokerins=smoker1ins_avg, nonsmokerins=nonsmoker1ins_avg, diff = diff1))
    print("")

    # Group 2: 36-45 years old
    smoker2 = 0
    nonsmoker2 = 0

    smoker2ins = 0
    nonsmoker2ins = 0

    for i in range(0,len(dataset[2])):
        if dataset[2][i]['smoker'] == 'no':
            nonsmoker2 += 1
            nonsmoker2ins += float(dataset[2][i]['charges'])
        else:
            smoker2 += 1
            smoker2ins += float(dataset[2][i]['charges'])

    smoker2pct = round(smoker2/len(dataset[2])*100, 2)
    nonsmoker2pct = round(nonsmoker2/len(dataset[2])*100, 2)

    smoker2ins_avg = round(smoker2ins/smoker2, 2)
    nonsmoker2ins_avg = round(nonsmoker2ins/nonsmoker2, 2)
    diff2 = round(smoker2ins_avg - nonsmoker2ins_avg, 2)

    print("There are {smoker}({smokerpct}%) smokers and {nonsmoker}({nonsmokerpct}%) non-smokers in people between the ages of 36 and 45.".format(smoker=smoker2, smokerpct=smoker2pct, nonsmoker=nonsmoker2, nonsmokerpct=nonsmoker2pct))
    print("In people between 36 and 45 years old, the average insurance cost for smokers is ${smokerins} and for non-smokers is ${nonsmokerins} (${diff} difference).".format(smokerins=smoker2ins_avg, nonsmokerins=nonsmoker2ins_avg, diff = diff2))
    print("")

    # Group 3: 46-55 years old
    smoker3 = 0
    nonsmoker3 = 0

    smoker3ins = 0
    nonsmoker3ins = 0

    for i in range(0,len(dataset[3])):
        if dataset[3][i]['smoker'] == 'no':
            nonsmoker3 += 1
            nonsmoker3ins += float(dataset[3][i]['charges'])
        else:
            smoker3 += 1
            smoker3ins += float(dataset[3][i]['charges'])

    smoker3pct = round(smoker3/len(dataset[3])*100, 2)
    nonsmoker3pct = round(nonsmoker3/len(dataset[3])*100, 2)

    smoker3ins_avg = round(smoker3ins/smoker3, 2)
    nonsmoker3ins_avg = round(nonsmoker3ins/nonsmoker3, 2)
    diff3 = round(smoker3ins_avg - nonsmoker3ins_avg, 2)

    print("There are {smoker}({smokerpct}%) smokers and {nonsmoker}({nonsmokerpct}%) non-smokers in people between the ages of 46 and 55.".format(smoker=smoker3, smokerpct=smoker3pct, nonsmoker=nonsmoker3, nonsmokerpct=nonsmoker3pct))
    print("In people between 46 and 55 years old, the average insurance cost for smokers is ${smokerins} and for non-smokers is ${nonsmokerins} (${diff} difference).".format(smokerins=smoker3ins_avg, nonsmokerins=nonsmoker3ins_avg, diff = diff3))
    print("")

    # Group 4: 56-65 years old
    smoker4 = 0
    nonsmoker4 = 0

    smoker4ins = 0
    nonsmoker4ins = 0

    for i in range(0,len(dataset[4])):
        if dataset[4][i]['smoker'] == 'no':
            nonsmoker4 += 1
            nonsmoker4ins += float(dataset[4][i]['charges'])
        else:
            smoker4 += 1
            smoker4ins += float(dataset[4][i]['charges'])


    smoker4pct = round(smoker4/len(dataset[4])*100, 2)
    nonsmoker4pct = round(nonsmoker4/len(dataset[4])*100, 2)

    smoker4ins_avg = round(smoker4ins/smoker4, 2)
    nonsmoker4ins_avg = round(nonsmoker4ins/nonsmoker4, 2)
    diff4 = round(smoker4ins_avg - nonsmoker4ins_avg, 2)

    print("There are {smoker}({smokerpct}%) smokers and {nonsmoker}({nonsmokerpct}%) non-smokers in people over 55 years old".format(smoker=smoker4, smokerpct=smoker4pct, nonsmoker=nonsmoker4, nonsmokerpct=nonsmoker4pct))
    print("In people older than 55 years old, the average insurance cost for smokers is ${smokerins} and for non-smokers is ${nonsmokerins} (${diff} difference).".format(smokerins=smoker4ins_avg, nonsmokerins=nonsmoker4ins_avg, diff = diff4))
    print("")

smoking(age_groups)    

#####################################################################################################################
#####################################################################################################################


    
    
    
```

## Part 3: Within different age groups and overall, does smoking correlate to higher BMI?
#### Note: No statistical tests were run to determine correlation.

```python
def bmi_list(list_dataset):
    
    smoker_count = 0
    nonsmoker_count = 0
    
    smoker_bmi = 0
    nonsmoker_bmi = 0
    
    for i in range(0, len(list_dataset)):
        if list_dataset[i]['smoker'] == 'no':
            nonsmoker_count += 1
            nonsmoker_bmi += float(list_dataset[i]['bmi'])
        else:
            smoker_count += 1
            smoker_bmi += float(list_dataset[i]['bmi'])
    
    smoker_bmi_avg = round(smoker_bmi/smoker_count, 2)
    nonsmoker_bmi_avg = round(nonsmoker_bmi/nonsmoker_count)
    print("The overall average BMI for smokers is {smokerbmi} and the average BMI for non-smokers is {nonsmokerbmi}.".format(smokerbmi=smoker_bmi_avg, nonsmokerbmi=nonsmoker_bmi_avg))
    
bmi_list(insurance_full)

def bmi_dict(dict_dataset):
    for k in range(0,len(dict_dataset)-1):
        smoker_count = 0
        nonsmoker_count = 0
        
        smoker_bmi = 0
        nonsmoker_bmi = 0
        
        for v in range(0, len(dict_dataset[k])):
            if dict_dataset[k][v]['smoker'] == 'no':
                nonsmoker_count += 1
                nonsmoker_bmi += float(dict_dataset[k][v]['bmi'])
            else:
                smoker_count += 1
                smoker_bmi += float(dict_dataset[k][v]['bmi'])
        if smoker_count != 0 and nonsmoker_count != 0:        
            smoker_bmi_avg = round(smoker_bmi/smoker_count, 2)
            nonsmoker_bmi_avg = round(nonsmoker_bmi/nonsmoker_count,2)
        print("In Group {k}, the average BMI for smokers {smokerbmi} and for non-smokers is {nonsmokerbmi}.".format(k=k, smokerbmi=smoker_bmi_avg, nonsmokerbmi=nonsmoker_bmi_avg))
            
bmi_dict(age_groups) 
        
```
